# Deep Agent — пример использования через Python SDK

Перед запуском:
1. Собран образ (`docker compose build`) и заполнен `.env` (см. `README_DEPLOY.md`).
2. PDF-файлы лежат в `./input/`.
3. Ниже укажите абсолютный путь к этой папке (`deploy/`) в `DEEP_AGENT_PROJECT`.

SDK — это тонкая обёртка над `docker compose run`: сам прогон идёт в контейнере.

```python
import os

# Путь к папке deploy/ (где docker-compose.yml и .env)
os.environ["DEEP_AGENT_PROJECT"] = "/абсолютный/путь/к/deploy"

from SDK import create_agent, load_agent
```

```python
# Создать агента один раз (инструкции сидятся из instructions/).
# schema_file/tools_file опциональны: по умолчанию — встроенная схема и штатные инструменты.
agent = create_agent(
    name="charter",
    business_rules="agent_init/buisness_rules.md",
    process="instructions/process.md",
    tool_tips="instructions/tool_tips.md",
    overwrite=True,
)
agent = load_agent("charter")  # повторно открыть позже
```

```python
# Прогнать один файл (путь относителен /workspace внутри контейнера)
result = agent.run("input/устав.pdf")

print(result.output_dir)   # ./output/charter/<timestamp>/
print(result.output[:800]) # markdown-отчёт
```

```python
# Главный результат — 7 полей в JSON
import json
print(json.dumps(result.structured, ensure_ascii=False, indent=2))
```

```python
# Пакетный режим: обработать всю папку input/ за один прогон
result = agent.run("input")
print(result.output_dir)
```
